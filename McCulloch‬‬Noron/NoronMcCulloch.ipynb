```python
! pip install pandas 
import pandas as pd

class NoronMcCulloch:
  def __init__(self):
    self._denderits = []
    self._summitionValue = 0

  def addDenderit(self,value,weight):
    self._denderits.append({
        value:value,
        weight:weight
    })
    self._summitionValue += value*weight
  def makeNose(self,activitionFucntion):
    return activitionFucntion(self._summitionValue);
  def getSum(self):
    return self._summitionValue

threshHolds = [1.5,-2.5,1,-1.5,0]
weightMatrix = [
    [2,-3,1,0],
    [-2,-2,-2,-2],
    [2,-1,1.5,-5],
    [1,1,1,1],
    [0.5,-0.5,0,-1],
]
data = []
for threshHold,weights in zip(threshHolds,weightMatrix):
  record = {} 
  for i in range(16):
    bits = f'{i:04b}'

    noron = NoronMcCulloch();

    noron.addDenderit(int(bits[0]),weights[0]);
    noron.addDenderit(int(bits[1]),weights[1]);
    noron.addDenderit(int(bits[2]),weights[2]);
    noron.addDenderit(int(bits[3]),weights[3]);

    isFeasible = noron.makeNose(lambda x: x>threshHold)

    record[i] = isFeasible;
  data.append(record)

df = pd.DataFrame(data,index=[
    str(weightMatrix[0])+','+str(threshHolds[0]),
    str(weightMatrix[1])+','+str(threshHolds[1]),
    str(weightMatrix[2])+','+str(threshHolds[2]),
    str(weightMatrix[3])+','+str(threshHolds[3]),
    str(weightMatrix[4])+','+str(threshHolds[4]),
]);
df.to_csv('out.csv')
display(df)
```
